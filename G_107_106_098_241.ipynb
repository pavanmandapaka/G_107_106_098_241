```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
# Fit on the Attrition column and transform in place
df['Attrition'] = le.fit_transform(df['Attrition'])
# Now 'No' â†’ 0, 'Yes' â†’ 1
print(df['Attrition'].value_counts())
```

```python
y = df['Attrition']
X = df.drop(columns='Attrition')
```

```python
numeric_cols     = X.select_dtypes(include=['int64','float64']).columns.tolist()
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()
print("Numeric:", numeric_cols)
print("Categorical:", categorical_cols)
```

```python
X_encoded = pd.get_dummies(
    X,
    columns=categorical_cols,
    drop_first=True
)
print("Shape after encoding:", X_encoded.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_encoded[numeric_cols] = scaler.fit_transform(X_encoded[numeric_cols])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
print("Train:", X_train.shape, "Test:", X_test.shape)
```

```python
print("Train shape:", X_train.shape, y_train.shape)
print("Test  shape:", X_test.shape,  y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, roc_auc_score, precision_recall_curve, auc


lr = LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42)
lr.fit(X_train, y_train)


y_pred_lr   = lr.predict(X_test)
y_proba_lr  = lr.predict_proba(X_test)[:,1]


f1_lr       = f1_score(y_test, y_pred_lr)
roc_auc_lr  = roc_auc_score(y_test, y_proba_lr)

precision, recall, _ = precision_recall_curve(y_test, y_proba_lr)
pr_auc_lr  = auc(recall, precision)

print(f"Logistic Regression â€” F1: {f1_lr:.3f}, ROC AUC: {roc_auc_lr:.3f}, PR AUC: {pr_auc_lr:.3f}")
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(
    max_depth=5,
    class_weight='balanced',
    random_state=42
)
dt.fit(X_train, y_train)

y_pred_dt  = dt.predict(X_test)
y_proba_dt = dt.predict_proba(X_test)[:,1]

f1_dt      = f1_score(y_test, y_pred_dt)
roc_auc_dt = roc_auc_score(y_test, y_proba_dt)

precision_dt, recall_dt, _ = precision_recall_curve(y_test, y_proba_dt)
pr_auc_dt  = auc(recall_dt, precision_dt)

print(f"Decision Tree â€” F1: {f1_dt:.3f}, ROC AUC: {roc_auc_dt:.3f}, PR AUC: {pr_auc_dt:.3f}")
```

```python
from sklearn.svm import SVC

svm = SVC(
    kernel='rbf',
    class_weight='balanced',
    probability=True,
    random_state=42
)
svm.fit(X_train, y_train)

y_pred_svm  = svm.predict(X_test)
y_proba_svm = svm.predict_proba(X_test)[:,1]

f1_svm      = f1_score(y_test, y_pred_svm)
roc_auc_svm = roc_auc_score(y_test, y_proba_svm)

precision_svm, recall_svm, _ = precision_recall_curve(y_test, y_proba_svm)
pr_auc_svm  = auc(recall_svm, precision_svm)

print(f"SVM â€” F1: {f1_svm:.3f}, ROC AUC: {roc_auc_svm:.3f}, PR AUC: {pr_auc_svm:.3f}")
```

```python
import pandas as pd

results = pd.DataFrame({
    'Model': ['Logistic Regression', 'Decision Tree', 'SVM'],
    'F1-score': [f1_lr, f1_dt, f1_svm],
    'ROC AUC':  [roc_auc_lr, roc_auc_dt, roc_auc_svm],
    'PR AUC':   [pr_auc_lr, pr_auc_dt, pr_auc_svm]
})

print(results)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve


plt.figure(figsize=(6,5))
for proba, label in zip([y_proba_lr, y_proba_dt, y_proba_svm],
                        ['Logistic', 'Tree', 'SVM']):
    fpr, tpr, _ = roc_curve(y_test, proba)
    auc_score   = roc_auc_score(y_test, proba)
    plt.plot(fpr, tpr, label=f"{label} (AUC={auc_score:.2f})")

plt.plot([0,1], [0,1], 'k--', linewidth=1)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves')
plt.legend()
plt.show()


plt.figure(figsize=(6,5))
for proba, label in zip([y_proba_lr, y_proba_dt, y_proba_svm],
                        ['Logistic', 'Tree', 'SVM']):
    precision, recall, _ = precision_recall_curve(y_test, proba)
    pr_auc = auc(recall, precision)
    plt.plot(recall, precision, label=f"{label} (AUC={pr_auc:.2f})")

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curves')
plt.legend()
plt.show()
```

```python

df['CurrentAnnualSalary'] = df['MonthlyIncome'] * 12

df['FutureAnnual_Fixed'] = df['CurrentAnnualSalary'] * (1 + growth_rate)
```

```python

def performance_increment(rating):
    if rating >= 4:
        return 0.10
    else:
        return 0.05

df['Increment_Pct'] = df['PerformanceRating'].apply(performance_increment)
```

```python
df['FutureAnnual_Perf'] = df['CurrentAnnualSalary'] * (1 + df['Increment_Pct'])
```

```python

df['P_stay'] = 1 - proba_leave

threshold = 0.6
df_stay = df[df['P_stay'] > threshold].copy()

print(f"Employees likely to stay (> {threshold*100:.0f}%):", df_stay.shape[0])
```

```python


X_reg = X_encoded.loc[df_stay.index]
y_reg = df_stay['FutureAnnual_Perf']

print("Regression dataset shape:", X_reg.shape, y_reg.shape)
```

```python
from sklearn.model_selection import train_test_split

Xr_train, Xr_test, yr_train, yr_test = train_test_split(
    X_reg, y_reg,
    test_size=0.2,
    random_state=42
)

print("Train:", Xr_train.shape, "Test:", Xr_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=100,
    max_depth=10,
    random_state=42
)
rf.fit(Xr_train, yr_train)
yr_pred_rf = rf.predict(Xr_test)
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1.0, random_state=42)
ridge.fit(Xr_train, yr_train)
yr_pred_ridge = ridge.predict(Xr_test)
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.1, random_state=42, max_iter=10000)
lasso.fit(Xr_train, yr_train)
yr_pred_lasso = lasso.predict(Xr_test)
```

```python
from sklearn.svm import SVR

svr = SVR(kernel='rbf', C=1.0, gamma='scale')
svr.fit(Xr_train, yr_train)
yr_pred_svr = svr.predict(Xr_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_percentage_error

def evaluate(y_true, y_pred, name):
    r2   = r2_score(y_true, y_pred)

    rmse = mean_squared_error(y_true, y_pred)**0.5
    mape = mean_absolute_percentage_error(y_true, y_pred)
    print(f"{name} â†’ RÂ²: {r2:.3f}, RMSE: {rmse:.2f}, MAPE: {mape:.3%}")

evaluate(yr_test, yr_pred_rf,    "Random Forest")
evaluate(yr_test, yr_pred_ridge, "Ridge")
evaluate(yr_test, yr_pred_lasso, "Lasso")
evaluate(yr_test, yr_pred_svr,   "SVR")
```
